```python
import pandas as pd
```

```python
sentiment_df = pd.read_csv('final_sentiment_results.csv', encoding='utf-8')
phone_df = pd.read_csv('new_data_cleaned.csv', encoding='utf-8')
```

```python
merged_df = pd.merge(phone_df, sentiment_df, on='phone_name', how='left')
```

```python
def price_retrieval(df, min_price, max_price):
    return df[(df['price'] >= min_price) & (df['price'] <= max_price)]

price_retrieval(merged_df, 2000000, 3000000)
```

```python
def price_sentiment_retrieval(df, min_price, max_price, top_k):
    temp = df[(df['price'] >= min_price) & (df['price'] <= max_price)]
    return temp.sort_values(by='sentiment_score', ascending=False).head(top_k)
price_sentiment_retrieval(merged_df, 2000000, 3000000, 10)
```
